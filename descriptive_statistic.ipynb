```python
data1=pd.read_csv('iris.csv')
```

```python
data1.head()
```

```python
data1.rename(columns={'SepalLengthCm':'SL','SepalWidthCm':'SW','PetalLengthCm':'PL','PetalWidthCm':'PW'},inplace=True)
data1.head()
```

```python
iris_setosa=data1[data1['Species']=='Iris-setosa']
iris_versicolor=data1[data1['Species']=='Iris-versicolor']
iris_virginica=data1[data1['Species']=='Iris-virginica']
```

1.Mean

```python
#sepal width

print(np.mean(iris_setosa['SW']))
print(np.mean(iris_versicolor['SW']))
print(np.mean(iris_virginica['SW']))
```

```python
#sepal length

print(np.mean(iris_setosa['SL']))
print(np.mean(iris_versicolor['SL']))
print(np.mean(iris_virginica['SL']))


print("You can do same with petal legth and petal width")
```

2.Median

```python
#petal Length

print(np.median(iris_setosa['PL']))
print(np.median(iris_versicolor['PL']))
print(np.median(iris_virginica['PL']))
```

3.Mode

```python
from scipy import stats

print(stats.mode(iris_setosa['PL']))
print(stats.mode(iris_virginica['PL']))
print(stats.mode(iris_versicolor['PL']))
```

unit-2: PDF AND CDF 


```python
sns.distplot(iris_setosa['PL'])
```

```python
sns.FacetGrid(data1,hue='Species',height=6).map(sns.distplot,'PL').add_legend()
```

Pdf= frequency/no of sample


Cdf=    FX(x) = P(X<=x)
        FX(x) =	function of X
        X	=	real value variable
        P	=	probability that X will have a value less than or equal to x


```python
counts, bin_edge= np.histogram(iris_setosa['PL'],bins=10)
```

```python
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
print(pdf)
```

```python
plt.plot(bin_edge[1:],pdf)
plt.plot(bin_edge[1:],cdf)
```
